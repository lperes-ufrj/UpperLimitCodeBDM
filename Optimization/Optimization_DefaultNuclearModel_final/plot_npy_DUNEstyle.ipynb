```python
#function to read the list of txt files
def read_eff_file(infile):
    """
    Read new Eff_Bkg_index_00a.txt files with columns:
      Mode SigStrength NegCosCut PosCosCut bestEff BkgExpect BkgError CutValue
    """

    rows = []
    with open(infile, "r") as f:
        for ln in f:
            ln = ln.strip()
            if not ln:
                continue
            if ln.lower().startswith("mode"):
                continue
            parts = ln.split()
            # expected: 8 columns (Mode + 7 floats)
            if len(parts) < 8:
                continue
            try:
                negcoscut = float(parts[2])  # NegCosCut
                poscoscut = float(parts[3])  # PosCosCut
                eff = float(parts[4])  # bestEff
                bkg = float(parts[5])  # BkgExpect
                cutvalue = float(parts[7])  # CutValue
            except ValueError:
                continue
            rows.append([negcoscut, poscoscut,eff, bkg,cutvalue])

        arr = np.asarray(rows, dtype=float)
    return arr
```

```python
def custom_sort_key(filename):
    """
    Sort by: gamma priority first (b1p1b1p5b10), then mass (m05m40)
    """
    import re
    
    # Extract mass value (m05, m10, m20, m40)
    mass_match = re.search(r'm(\d+)', filename)
    mass = int(mass_match.group(1)) if mass_match else 999
    
    # Extract gamma/beta name (b10, b1p1, b1p5)
    gamma_match = re.search(r'b(1p\d|10)', filename)
    gamma_str = gamma_match.group(1) if gamma_match else 'zzz'
    
    # Gamma priority: b1p1=0, b1p5=1, b10=2
    gamma_priority = {'1p1': 0, '1p5': 1, '10': 2}.get(gamma_str, 999)
    
    return (gamma_priority, mass)

# Sort all three lists with custom key
CosSunSignalAfterCut_npy_list = sorted(CosSunSignalAfterCut_npy_list, key=custom_sort_key)
CosSunAtmAfterCut_npy_list = sorted(CosSunAtmAfterCut_npy_list, key=custom_sort_key)
#CVN_npy_list_allevts = sorted(CVN_npy_list[:1], key=custom_sort_key)

# Also create sorted eff_file indices 
sorted_indices = sorted(range(len(CosSunSignalAfterCut_npy_list)), 
                        key=lambda i: custom_sort_key(CosSunSignalAfterCut_npy_list[i]))


print("Sorted Signal files:")
for f in CosSunSignalAfterCut_npy_list:
    print(f)
print("\nSorted Atm files:")
for f in CosSunAtmAfterCut_npy_list:
    print(f)
for f in eff_file:
    print(f)
```
